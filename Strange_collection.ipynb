Данная совокупность уровнений сводится к поиску всех возможных двух множителей:


```math
a = xy;
b = xy;
c = xy;
```

> Для того чтобы в случае, если на вход придет число приближенное к концу диапазона n = 10^9, не перебирать весь диапазон, можно ограничить перебор диапазона от 1 до n^0.5
---

- Данное решение выдает валидный результат, но недостаточно оптимизированно по времени (контест фейлит на времени решения), текущая алгоритмитическая сложность O(n**0,5). 

````python
from math import sqrt


def mnoj(a,res):
  for x in range(1, int(sqrt(a)+1)):
    if not (a % x):
      res.append(x)
      res.append(a//x)

def mnojABC(a,b,c):
  res = []
  mnoj(a,res)
  mnoj(b,res)
  mnoj(c,res)
  res = set(res)
  return len(res)


t = int(input())
result = []
for i in range(0, t):
  abc = [int(i) for i in input().split()]
  result.append(mnojABC(abc[0],abc[1],abc[2]))
  
for i in result:
  print(i)
````
